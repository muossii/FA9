```python
music_D['Likes'] = music_D['Likes'].fillna(0)
music_D['Comments'] = music_D['Comments'].fillna(0)

music_D.dropna(inplace=True)

music_D

```

```python
#Describe (descriptive measures: count, mean, std, min, 25%, 50%, 75%, max) the cleaned data.
music_D.describe()
```

**Exploring Album Type** 

```python
count_albumt = music_D['Album_type'].value_counts()
print(count_albumt)
```

```python
#Create a pie chart for Album Types as a Title and make some remarks.

labels = count_albumt.index.tolist()
sizes = count_albumt.values.tolist()
mpl.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

mpl.title('Album Types')
mpl.legend(labels, loc = 'best')
mpl.show()
```

**The pie chart represents the poularity of album types, or music releases. From the data, we can conclude that:**
- Album type is the most popular, having a percentage of 72.0.
- Singles are more popular than compilation, however, albums are still popular than the latter. From here, we can say that listeners prefer individual releases than listening to compilations of songs from different artists. 

```python
#Analyze songs by artist and views-youtube, stream-spotify.

grouped_artists = music_D.groupby('Artist')[['Views', 'Stream']].sum()
sorted_artists = grouped_artists.sort_values(['Views', 'Stream'], ascending=False)

top_ten = sorted_artists.head(10)

top_ten
```

```python
#Create bar charts for Top 10 Artists on Youtube and Top 10 Artists on Spotify in 1 row of display.

music_D_views = music_D.groupby('Artist')['Views'].sum().sort_values(ascending=False)[:10]
music_D_streams = music_D.groupby('Artist')['Stream'].sum().sort_values(ascending=False)[:10]

fig, (ax1, ax2) = mpl.subplots(1,2, figsize = (12,6))


#top 10 yt artists
ax1.set_title('Top Ten Artists on Youtube')
music_D_views.plot(kind='bar', ax = ax1)


#top 10 spotify artists
ax2.set_title('Top Ten Artists on Spotify')
music_D_streams.plot(kind='bar', ax = ax2)

ax1.set_xlabel('Artist')
ax1.set_ylabel('Views')

ax2.set_xlabel('Artist')
ax2.set_ylabel('Streams')
fig.tight_layout()
mpl.show()
```

Figure 1 (left) represents the top ten most viewed artists on youtube, while Figure 2 (right) represents the top ten most streamed artists on Spotify.

**From both graphs, the following are observed:**

Ed sheeran is both the most viewed and streamed artist with 1.5 billions views on Youtube, while he came second on Spotify. Cocomelon and Katy Perry came in second place and third place, respectively, while Charlie Puth being the fourth place and Luis Fonsi on 5th place.

On the other hand, Post Malone is the first streamed artist on Spotify, followed by Ed Sheeran and Dua lipa, who came second and third respectfully.

```python
#Determine the top 10 artists on the Spotify platform whose songs have more likes and display a bar chart with a title Number of
#Likes per Artist

likes_Artist = music_D.groupby(['Artist'])[['Likes']].sum().sort_values(by=['Likes'], ascending=False).head(10)

likes_Artist = likes_Artist.reset_index()

likes_Artist.plot.bar(x='Artist',y='Likes',title='Number of Likes per Artist')
mpl.ylabel('Likes per Artist')
```

BTS has the most likes, followed closely by BlackPink, these 2 are the only artists to have surpassed the 1e8 mark. 
they are followed by Charlie Puth, Ed Sheeran, Justin Bieber, Billie Eilish, Imagine Dragons, Alan Walker, DJ Snake, Luis Fonsi
the last 8 has likes around 8e7-7e7 

```python
#Explore single releases by artists and identify top performers based on likes and display a table for each artist with their 
#Views and Likes as the other columns.

singles = music_D[music_D.Album_type == 'single']
grouped = singles.groupby("Artist").agg({ "Likes": "sum","Views": "sum"})

top_Artist_bySingles = grouped.sort_values("Likes", ascending=False).head(10)

top_Artist_bySingles
```

BLACKPINK has the most likes as an artist based on their singles, as well as the most views on her singles. They are followed by
Wiz Khalifa, who has the second most likes on singles, albeit almost half of BlackPink's, as well as second most views.
the ranking for likes is almost the same as the ranking for views, with a few exceptions in marshmello, selena, and ozuna. Tanishk Bagchi had the lowest likes of the ten, but Bizarrap had the lowest views of the ten.

```python
#Exploring compilation releases by artists and identify top performers based on views and likes and display a table for each 
#artist with their Views and Likes as the other columns. 

couples = music_D[music_D.Album_type == 'compilation']
grouped_couples = couples.groupby("Artist").agg({ "Likes": "sum","Views": "sum"})

top_Artist_Compilation = grouped_couples.sort_values(["Likes"], ascending=False).head(10)

top_Artist_Compilation
```

Shakira has the most likes as well as the most views on her compilation releases. She is then followed by Ellie Goulding.
Since the 3rd place, the ranking sorted by likes stops being equivalent to the rank sorted by views, e.g. Earth, Wind and Fire has more likes than DJ Luian, Mambo kingz, and PSY, but they have less views on their compilation releases than the latter 3. Dave Stewart had the least likes as well as the least views in the top 10 sorted by likes.

```python
#Identify the top 10 channels based on views and display a bar chart of it and make some remarks.
#Channel
views_Channel = music_D.groupby(['Channel'])[['Views']].sum().sort_values(by=['Views'], ascending=False).head(10).reset_index()

views_Channel.plot.bar(y='Views',x='Channel',title='Views per Channel')
mpl.xlabel('Channel')
mpl.ylabel('Views')
```

T-Series has the most views per channel, with more than half than the second place, LuisFonsiVEVO. From then on, the change in views per channel is more gradual. The Zee Music Company had the least views out of the 10.
